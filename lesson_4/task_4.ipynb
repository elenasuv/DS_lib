```python
import pandas as pd
```

```python
a = pd.read_csv('creditcard.csv')
a.iloc[:10, :]
```

```python
plt.style.use('fivethirtyeight')
```

```python
b = a['Class'].value_counts()
b
```

```python
b.plot(kind='bar')
plt.show()
```

```python
b.plot(kind='bar')
plt.yscale(value="log")
plt.show()
```

```python
c = a.loc[(a['Class'] == 0), 'V1']
c.reset_index(drop=True, inplace=True)
c
```

```python
d = a.loc[(a['Class'] == 1), 'V1']
d.reset_index(drop=True, inplace=True)
d
```

```python
plt.hist(c, bins=20, alpha=0.5, color='grey')
plt.hist(d, bins=20, alpha=0.5, color='red')
plt.yscale(value="log")
plt.xlabel('Class')
plt.legend(labels=['Class 0', 'Class 1'])
plt.show()
```
